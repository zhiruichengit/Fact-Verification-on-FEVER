```python
import pandas as pd
import fever_data
from fever_data import read_fever_examples, read_blind_fever_examples, read_wiki_pages
```

```python
label_lookup = {
    0: "SUPPORTS",
    1: "REFUTES",
    2: "NOT ENOUGH INFO",
}

train_exs = read_fever_examples("data/train.jsonl")
dev_exs = read_fever_examples("data/shared_task_dev.jsonl")
test_exs = read_blind_fever_examples("data/shared_task_test.jsonl")

train = pd.DataFrame([{"claim": ex.claim, "label": label_lookup[ex.label], "num_tokens": len(ex.words)} for ex in train_exs])
dev = pd.DataFrame([{"claim": ex.claim, "label": label_lookup[ex.label], "num_tokens": len(ex.words)} for ex in dev_exs])
test = pd.DataFrame([{"claim": ex.claim, "num_tokens": len(ex.words)} for ex in test_exs])

wiki_pages = read_wiki_pages("data/fever_wiki_pages")
wiki_split = wiki_pages["wikipedia_pages"] if hasattr(wiki_pages, "keys") else wiki_pages
```

```python
print("train shape:", train.shape)
print("dev shape:", dev.shape)
print("test shape:", test.shape)
train.head()
```

```python
print("===== train labels =====")
print(train["label"].value_counts())
print("===== dev labels =====")
print(dev["label"].value_counts())
```
